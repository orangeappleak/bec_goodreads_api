```python
import requests
from bs4 import BeautifulSoup as bs
url= "https://www.goodreads.com/shelf/show/psychology"
page = requests.get(url) 
soup = bs(page.content, 'html.parser')
#print(soup)
```

```python
titles = soup.find_all('a', class_='bookTitle')
authors = soup.find_all('a', class_='authorName')
ratings = soup.find_all('span', class_='greyText')
images = soup.find_all('a', class_='leftAlignedImage')
#print(titles)

book_title = []
for tag in titles:
    book_title.append(tag.text)
book_title

book_author = []
for tag in authors:
    book_author.append(tag.text)
book_author

book_rating = []
for tag in ratings:
    book_rating.append(tag.text)
book_rating

titles = soup.find_all('a', {'class' : 'bookTitle'})
#len(titles)

title_urls = []
base_url = 'https://www.goodreads.com/shelf/show/psychology'
for tag in titles:
    title_urls.append(base_url + tag['href'])
#title_urls

authors = soup.find_all('a', {'class' : 'authorName'})
#len(authors)

author_urls = []
for tag in authors:
    author_urls.append(tag['href'])
#author_urls


```

```python
import pandas as pd

topics_dict = {
    'title' :  book_title,
    'author' : book_author,
    'title_url' : title_urls,
    'author_url' : author_urls
}
topics_df = pd.DataFrame(topics_dict)
topics_df
```

```python
topics_df.to_csv('topics.csv', index=None)
```
