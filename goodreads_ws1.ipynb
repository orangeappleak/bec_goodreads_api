```python
import requests
from bs4 import BeautifulSoup as bs4

pages=[]

pages_to_scrape=5
for i in range(1,pages_to_scrape+1):
    url= ("https://www.goodreads.com/list/show/21995.Best_21st_Century_Non_Fiction?page={}").format(i)
    pages.append(url)
#print(pages)
for item in pages:
    page = requests.get(item) 
    soup = bs4(page.content, 'html.parser')
    #print(soup)
```
